```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('data.csv')
df = df[df['year'] >= 2000]
df = df.reset_index(drop=True)
df
```

```python
features = ['danceability', 'loudness', 'speechiness','valence','acousticness','energy','instrumentalness','liveness','tempo']  
X = df[features]
X
```

```python
import h5py
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def compute_and_save_cosine_similarity_h5(X, batch_size=1000, output_file='cosine_sim_matrix.h5'):
    num_samples = X.shape[0]
    
    with h5py.File(output_file, 'w') as f:
        dset = f.create_dataset('cosine_similarity', (num_samples, num_samples), dtype=np.float32, compression='gzip', compression_opts=9)
        
        for i in range(0, num_samples, batch_size):
            end_i = min(i + batch_size, num_samples)
            for j in range(i, num_samples, batch_size):
                end_j = min(j + batch_size, num_samples)
                batch_i = X[i:end_i]
                batch_j = X[j:end_j]
                
                cos_sim_batch = cosine_similarity(batch_i, batch_j)
                
                dset[i:end_i, j:end_j] = cos_sim_batch

compute_and_save_cosine_similarity_h5(X)
```

```python
df_search=df.copy()
df_search['name'] = df['name'] + ' ' + df['artists']
df_search.head()
```

```python
import pandas as pd
import difflib

def get_closest_song(song_name, df):
    song_name = song_name.lower().strip()
    df['normalized_name'] = df['name'].str.lower().str.strip()
    list_of_all_titles = df['normalized_name'].tolist()
    find_close_matches = difflib.get_close_matches(song_name, list_of_all_titles)
    
    if not find_close_matches:
        print('No match found')
        return None
    
    print("\nSimilar songs:")
    similar_titles = df[df['normalized_name'].isin(find_close_matches)]
    
    for i, name in enumerate(find_close_matches):
        matching_rows = similar_titles[similar_titles['normalized_name'] == name]
        for index, row in matching_rows.iterrows():
            print(f"{i}: {row['name']} by {row['artists']}")
            break  
    
    try:
        choice = int(input(f"Enter your choice number (0 to {len(find_close_matches) - 1}): "))
        if 0 <= choice < len(find_close_matches):
            chosen_name = find_close_matches[choice]
            chosen_song = similar_titles[similar_titles['normalized_name'] == chosen_name]
            chosen_song_info = chosen_song[['name', 'artists']].iloc[0]
            return chosen_song_info
        else:
            print("Invalid choice")
            return None
    except ValueError:
        print("Invalid input. Please enter a number.")
        return None



song_name = input('Enter your favourite song name: ')
chosen_song_info = get_closest_song(song_name, df)
if chosen_song_info is not None:
    print(f"You selected: {chosen_song_info['name']} by {chosen_song_info['artists']}")
search_name=chosen_song_info['name'] +' '+ chosen_song_info['artists']
print(search_name)
```

```python
import h5py

with h5py.File('cosine_sim_matrix.h5', 'r') as f:
    cos_sim_matrix = f['cosine_similarity']
    matching_index = df_search[df_search['name'] == search_name].index[0]
    similarity_scores = cos_sim_matrix[matching_index:matching_index + 1]
    index_score_pairs = [(i, similarity_scores[0, i]) for i in range(similarity_scores.shape[1])]
    print(index_score_pairs)
```
